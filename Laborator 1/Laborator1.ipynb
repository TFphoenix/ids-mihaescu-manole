```python
# EX 4
def collatz_conjecture(number):
    collatz_list = [number]
    n = number
    while n != 1: 
        if n % 2 == 0:
            n //= 2
        else:
            n = n * 3 + 1
        collatz_list.append(n)
    return collatz_list
        
for number in range(1, 1001):
    print(f"Collatz Conjecture for {number} is: ", end = "")
    print(collatz_conjecture(number))
```

```python
# EX 7
def build_number(number):
    initial = str(number)
    final = str()
    
    even_count = 0
    odd_count = 0
    for digit in initial:
        if int(digit) % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
            
    final = str(even_count) + str(odd_count) + str(sum([even_count, odd_count]))
    return int(final)

def build_number_until_123(number):
    while number != 123:
        print(number, end = ", ")
        number = build_number(number)
    print(number)

for number in range(1, 1001):
    print(f"Result for {number} is: ", end = "")
    build_number_until_123(number)
```

```python
# EX 8
translator_en = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]

def translate_number(number, translator):
    number_str = str()
    for digit in str(number):
        number_str += translator[int(digit)]
    return number_str

def verify_number(number, translator):
    while number != 4:
        number_translation = translate_number(number, translator)
        print(f"{number}({number_translation}),", end = " ")
        number = len(number_translation)
    print(f"4({translator[4]})")        

def verify_language(translator):
    n = int(input("n: "))
    for i in range(1, n + 1):
        print(f"Verification for number {i}: ", end = "")
        verify_number(i, translator)
        
verify_language(translator_en)
```

```python
# EX 9
translator_ro = ["zero", "unu", "doi", "trei", "patru", "cinci", "sase", "sapte", "opt", "noua"]
translator_de = ["null", "einz", "zwei", "drei", "vier", "funf", "sechs", "sieben", "acht", "neun"]

verify_language(translator_ro)
verify_language(translator_de)
```

```python
# EX 10
def number_divisors(number):
    divisors = [1]
    for d in range(2,number):
        if number % d == 0:
            divisors.append(d)
    divisors.append(number)
    return divisors

def divisors_digit_sum(divisors):
    dd_sum = 0
    for divisor in divisors:
        for digit in str(divisor):
            dd_sum += int(digit)
    return dd_sum

n = int(input("n: "))
while n != 15:
    print(f"{n},", end = " ")
    n = divisors_digit_sum(number_divisors(n))
print(n)
```
